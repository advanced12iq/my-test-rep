```python
import pandas as pd

df = pd.read_csv('optimization_results.csv')
prev = None
for index, row in df.iterrows():
    if type(row['function']) is str:
        prev = row['function']
    df.iloc[index, df.columns.get_loc('function')] = prev
    if index+1 % 1000 == 0:
        display(f"{index}/{len(df)}") 
df
```

```python
df.to_csv('optimization_results.csv', index=False)
```
